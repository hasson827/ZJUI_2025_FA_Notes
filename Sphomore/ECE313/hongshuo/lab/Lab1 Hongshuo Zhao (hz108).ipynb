To specify a function, we need to use the "def" keyword. You need to give the number of inputs and have a return line to end your function. Below is a function that returns the factorial of the input.

```python
def factorial(x):
    c = 1
    for i in range(x,1,-1): #range(x,1,-1) creates a vector from x to 2 in -1 increments
        c = c*i
    return c

print (factorial(5))
```

You can also return multiple outputs. Technically, we are still returning a single object, but it is a tuple. We can unpack the tuple when we call the function. Below is a function that returns the first and last digit of any integer.

```python
def firstNlast(x):
    l = x%10 # Uses the modulus operator %
    while x>0:
        f = x%10
        x = int(x/10)
    return f,l

x = 70094921348
first, last = firstNlast(x)
print (first, last)
```

The returned items get returned as a tuple and you can individually retrieve them by setting them equal to another tuple.

### **Part 1: Events and Set Operations**
In probability theory, the **sample space** ($\Omega$) is the set of all possible outcomes of an experiment. An **outcome** ($\omega$) is an element of the sample space. An **event** is a subset of the sample space. For example, in the experiment of rolling a standard six-sided die, the sample space is $\Omega = \{1, 2, 3, 4, 5, 6\}$. The event "the number rolled is less than or equal to 3" can be represented by the set $A = \{1, 2, 3\}$.

We can perform set operations on events, such as **Union** (A $\cup$ E, the logical "OR"), **Intersection** (A $\cap$ E, the logical "AND"), and **Complement** ($E^c$, the logical "NOT").

**De Morgan's Law** is a fundamental principle in set theory that describes the relationship between these operations:
* $(A \cup B)^c = A^c \cap B^c$
* $(A \cap B)^c = A^c \cup B^c$

Python's `set` type is very effective for representing and manipulating events. Here is an example:

```python
p = 1./4 #Sets the probability, uses decimal to create double (not integer)
# Define the sample space and two events
Omega = {1, 2, 3, 4, 5, 6}
A = {1, 2, 3}
E = {2, 4, 6}

# Calculate the union (A U E)
union_AE = A.union(E)
print(f"A U E = {union_AE}")

# Calculate the intersection (A  E)
intersection_AE = A.intersection(E)
print(f"A  E = {intersection_AE}")

# Calculate the complement (A^c)
complement_A = Omega.difference(A)
print(f"A^c = {complement_A}")
```

### **Part 2: Axioms of Probability and Counting**

The theory of probability is built upon three fundamental axioms:
1.  **Axiom 1 (Non-negativity):** For any event A, $P(A) \ge 0$.
2.  **Axiom 2 (Normalization):** $P(\Omega) = 1$, the probability of the entire sample space is 1.
3.  **Axiom 3 (Additivity):** For any sequence of mutually exclusive events $A_1, A_2, \dots$, the probability of their union is the sum of their individual probabilities.

For a finite sample space where all outcomes are **equally likely**, the probability of an event A can be calculated as:
$P(A) = \frac{|A|}{|\Omega|}$, where $|A|$ is the number of outcomes in event A (the cardinality of the set), and $|\Omega|$ is the total number of outcomes in the sample space.

**The Fundamental Principle of Counting:** If an experiment can be described as a sequence of two steps, with $m$ ways to perform the first step and $n$ ways to perform the second, then there are $m \times n$ total possible outcomes. For example, when rolling two dice, there are $6 \times 6 = 36$ possible outcomes.

```python
# Example: Calculate the probability of rolling an even number with a single die
Omega_size = 6
E = {2, 4, 6}
E_size = len(E)

# Calculate the probability
prob_E = E_size / Omega_size
print(f"P(E) = {E_size}/{Omega_size} = {prob_E}")
```

### **Part 3: Derived Properties of Probability**

From the three axioms, we can derive several useful properties of probability. One of the most important is the **Additive Rule**, which applies to any two events A and B (they do not need to be mutually exclusive):

$P(A \cup B) = P(A) + P(B) - P(A \cap B)$

This formula can be proven by partitioning the event $A \cup B$ into three mutually exclusive sets: $A \cap B^c$, $A \cap B$, and $A^c \cap B$.

For three events A, B, and C, this rule can be extended (as shown in your Lecture 3 notes):
$P(A \cup B \cup C) = P(A) + P(B) + P(C) - P(A \cap B) - P(A \cap C) - P(B \cap C) + P(A \cap B \cap C)$
